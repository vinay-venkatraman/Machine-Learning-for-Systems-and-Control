```python
import numpy as np

def angle_normalize(x):

    return ((x + np.pi) % (2 * np.pi)) - np.pi
```

```python
import gymnasium as gym
import gym_unbalanced_disk
import numpy as np
import time
from stable_baselines3 import SAC

env = gym_unbalanced_disk.UnbalancedDisk_sincos(dt=0.025, umax=3.)

env.reward_fun =  lambda self: - (20 * angle_normalize(self.th - np.pi) ** 2 + 0.01 * (self.omega**2))

model = SAC(
    "MlpPolicy",
    env,
    ent_coef='auto_0.1',
    verbose=1,
)
model.learn(total_timesteps=100_000)
```

```python
model.save("sac_disk")
```

```python
import gymnasium as gym
import gym_unbalanced_disk
import numpy as np
import time
from stable_baselines3 import SAC


env = gym_unbalanced_disk.UnbalancedDisk_sincos(dt=0.025, umax=3.)

env.reward_fun =  lambda self: - (20 * angle_normalize(self.th - np.pi) ** 2 + 0.01 * (self.omega**2))

model = SAC.load("sac_disk", env=env, device="cpu")

obs, info = env.reset()
Rewards = []
Theta = []
Omega = []
Actions = []
try:
    for _ in range(2500):
        action, _states = model.predict(obs)
        obs, reward, terminated, truncated, info = env.step(action)
        Theta.append(obs[0])
        Omega.append(obs[1])
        Rewards.append(reward)
        Actions.append(action)
        env.render()
        time.sleep(1/24)
        if terminated or truncated:
            env.reset()
finally:
    env.close()
```
